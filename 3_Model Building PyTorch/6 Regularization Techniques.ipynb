4. **Regularization Techniques:**
   - **Purpose:** Regularization techniques are used to prevent overfitting by adding penalty terms to the loss function to discourage large weights or complex models.
   - **Mathematics:** Regularization techniques include L1 regularization (Lasso), L2 regularization (Ridge), and dropout.
   - **Mathematics of L2 Regularization:** L2 regularization adds a penalty term to the loss function:
     \[ L_{\text{reg}} = \lambda \sum_{i=1}^n W_i^2 \]
     where
     - \(L_{\text{reg}}\) is the regularization term.
     - \(\lambda\) is the regularization strength.
     - \(W_i\) are the model's weights
